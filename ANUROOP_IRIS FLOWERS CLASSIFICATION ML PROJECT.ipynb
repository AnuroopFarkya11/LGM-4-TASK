# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
iris_df.head()
```

```python
iris_df['species']=le.fit_transform(iris_df['species'])
iris_df.head()
```

# MACHINE LEARNING 

# TRAINING MODEL OF LOGISTIC REGRESSION

```python
X=iris_df.drop(columns=['species'])#INDEPENDENT VARIABLE
Y=iris_df['species']
```

```python
#SPLIT TRAIN AND TEST RECORD FOR TRAINING A MODEL
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=21)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
print("ACCURACY OF LINEAR REGRESSION MODEL IS : ",lr.score(x_test,y_test)*100)
```

# Verification for accuracy

```python
#MEAN SQUARED ERROR 
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

# TRAINING MODEL OF LOGISTIC REGRESSION

```python
P=iris_df.drop(columns=['species'])#INDEPENDENT VARIABLE
Q=iris_df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
p1_train,p2_test,q1_train,q2_test=train_test_split(P,Q,test_size=0.3)
```

```python
#Import model
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo=LogisticRegression()
```

```python
lo.fit(p1_train,q1_train)
```

```python
q2_pred=lo.predict(p2_test)
```

```python
print("ACCURACY OF LOGISTIC REGRESSION MODEL IS : ",lo.score(p2_test,q2_test)*100)
```

# Verification for accuracy

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(q2_test,q2_pred)
```

# TRAINING MODEL OF K-NEIGHBORS  CLASSIFIER

```python
K1=iris_df.drop(columns=['species'])#INDEPENDENT VARIABLE
K2=iris_df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
k1_train,k1_test,k2_train,k2_test=train_test_split(K1,K2,test_size=0.3)
```

```python
#IMPORT MODEL
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier()
```

```python
kn.fit(k1_train,k2_train)
```

```python
k2_pred=kn.predict(k1_test)
```

```python
print("ACCURACY OF KNeighbors Classifier MODEL IS : ",kn.score(k1_test,k2_test)*100)
```

# Verification for accuracy

```python
#MEAN SQUARED ERROR 
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(k2_test,k2_pred)
```
