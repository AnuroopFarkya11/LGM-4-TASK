## STEPS

The following are the steps to be performed:-

1.  Convert the RGB color image to grayscale.
2.  Invert the grayscale image to get negative.
3.  Apply Gaussian Blur to the negative from step 2.
4.  Invert the blurred image.








# Import the cv2 module

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

```python
img = cv2.imread("lgg2.jfif")
```

```python
cv2_imshow(img)
```

# RESIZE THE IMAGE 

```python
scale_percent=0.50
width=int(img.shape[1]*scale_percent)
height=int(img.shape[0]*scale_percent)
dimension=(width,height)
```

```python
resized=cv2.resize(img,dimension,interpolation=cv2.INTER_AREA)
```

```python
cv2_imshow(resized)
```

# Convert the image to gray scale

```python
gray_img = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(gray_img)
```

# Invert the gray image

```python
inverted_gray_img=cv2.bitwise_not(gray_img)
```

```python
cv2_imshow(inverted_gray_img)
```

# Blur the image using Gaussian function

```python
blurred_img=cv2.GaussianBlur(inverted_gray_img,(21,21),sigmaX=0, sigmaY=0)
```

```python
cv2_imshow(blurred_img)
```

# Invert the blurred image

Blend the gray-scale image from step 1 with the blurred negative from step 3 using a color dodge.

```python
def dodgev2(x,y):
  return cv2.divide(x, 255-y, scale=256)
```

```python
final_img= dodgev2(gray_img, blurred_img)
```

# SKETCH IS READY!!

```python
cv2_imshow(final_img)
```
